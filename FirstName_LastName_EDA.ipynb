```python
#  importing libaries for data analysis
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt   
import seaborn as sns    
```

```python
# Read the CSV files into dataframes
Customers_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\data science intern\Customers.csv")
transactions_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\data science intern\Transactions.csv")
products_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\data science intern\Products.csv")

# Display the first few rows of each dataframe to understand their structure
transactions_df.head(), products_df.head(), Customers_df.head()
```

```python
transactions_df.describe,Customers_df.describe,products_df.describe
```

```python
# getting info from the datasets
transactions_df.info(),Customers_df.info(),products_df.info()
```

```python
# merging the files
# Merge transactions with products
transactions_products = transactions_df.merge(products_df, on="ProductID", how="left")

# Merge the resulting DataFrame with customers
final_merged_df = transactions_products.merge(Customers_df, on="CustomerID", how="left")
final_merged_df

```

```python
print(final_merged_df.head())
print(final_merged_df.info())
```

```python
# 1. Top-Selling Products by Quantity
top_products = final_merged_df.groupby("ProductName")["Quantity"].sum().sort_values(ascending=False).head(5)
# Low-selling products by Quantity
Low_products = final_merged_df.groupby("ProductName")["Quantity"].sum().sort_values(ascending=False).tail(5)

# 2. Revenue by Category
revenue_by_category = final_merged_df.groupby("Category")["TotalValue"].sum().sort_values(ascending=False)

# 3. Transaction Trends- Monthly Sales
final_merged_df["TransactionDate"] = pd.to_datetime(final_merged_df["TransactionDate"])
final_merged_df["Month"] = final_merged_df["TransactionDate"].dt.to_period("M")
monthly_sales = final_merged_df.groupby("Month")["TotalValue"].sum()

# 4. Top Customers by Spending
top_customers = final_merged_df.groupby("CustomerID")["TotalValue"].sum().sort_values(ascending=False).head(5)

top_products, Low_products, revenue_by_category, monthly_sales.head(), top_customers
```

```python
monthly_sales.plot(kind="line", figsize=(10, 6), title="Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales (USD)")
plt.grid()
plt.show()
```
