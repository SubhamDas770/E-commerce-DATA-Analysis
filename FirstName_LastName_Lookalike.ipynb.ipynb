Performing Lookalike


```python
# importing libraies 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn  
```

```python
# Read the CSV files into dataframes
Customers_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\data science intern\Customers.csv")
transactions_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\data science intern\Transactions.csv")
products_df = pd.read_csv(r"C:\Users\Lenovo\Downloads\data science intern\Products.csv")

# Display the first few rows of each dataframe to understand their structure
products_df.head(), Customers_df.head()
```

```python
# merging the files
# Merge transactions with products
transactions_products = transactions_df.merge(products_df, on="ProductID", how="left")

# Merge the resulting DataFrame with customers
final_merged_df = transactions_products.merge(Customers_df, on="CustomerID", how="left")
final_merged_df

```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler
```

```python
features = final_merged_df.groupby("CustomerID").agg({
    "TotalValue": "sum",               # Total spending
    "Quantity": "sum",                 # Total quantity purchased
    "ProductID": "nunique"             # Number of unique products purchased
}).reset_index()
features
```

```python
#  Normalize the features
scaler = StandardScaler()
normalized_features = scaler.fit_transform(features.iloc[:, 1:])  # Exclude CustomerID
normalized_features
```

```python
# Compute cosine similarity
similarity_matrix = cosine_similarity(normalized_features)

# Find top 3 lookalikes for the first 20 customers
customer_ids = features["CustomerID"].values
lookalike_dict = {}
```

Selecting first 20 customer id and then mapping it


```python
for i, cust_id in enumerate(customer_ids[:20]):  # Only for the first 20 customers
    similarities = list(enumerate(similarity_matrix[i]))
    sorted_similarities = sorted(similarities, key=lambda x: x[1], reverse=True)[1:4]  # Exclude self
    lookalike_dict[cust_id] = [(customer_ids[j], round(score, 2)) for j, score in sorted_similarities]

# Step 5: Save the results to a CSV file
lookalike_df = pd.DataFrame({
    "CustomerID": lookalike_dict.keys(),
    "Recommendations": lookalike_dict.values()
})
```
